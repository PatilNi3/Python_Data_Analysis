# Pandas In One

```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df
```

```python
#Read headers
df.columns
```

```python
# Read each columns -> df.column_name or df['column_name']
df['Name']
```

```python
# Read multiple columns
df[['Name','Type 1', 'Type 2']]
```

```python
#Read rows
df['Name'][5]
```

```python
#Read rows
df['Name'][1:6]
```

```python
#Read rows using loc
df.loc[1]
```

```python
#Read rows using loc
df.loc[0:5]    #including:including
```

```python
#Read rows using iloc
df.iloc[1]
```

```python
#Read rows using iloc
df.iloc[0:5]    #including:excluding
```

```python
#To fetch specific like Charmander
df.iloc[4,1]
```

```python
#Fetching specific rows like whose Type1 = Electric
e = df[df['Type 1']=='Electric']
```

```python
e
```

```python
#Sorting dataframe
df.sort_values('Type 1')
```

```python
#Sorting dataframe
df.sort_values('Type 1', ascending=False)
```

```python
#Sorting dataframe
df.sort_values(['Type 1', 'Type 2'])
```

```python
#Sorting dataframe
df.sort_values(['Type 1', 'Type 2'], ascending=False)
```

```python
#Sorting dataframe
df.sort_values(['Type 1', 'Type 2'], ascending=[1,0])    # 1=ascending, 0=descending
```
